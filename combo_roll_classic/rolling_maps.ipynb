```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
import glob
from rolling_classic import make_rolling_footprints
```

```python
outdir = 'rolling_maps'
resultsDb = db.ResultsDb(outDir=outdir)
db_files = glob.glob('*nslice*.db')
names = [name.replace('_v1.6.1_10yrs.db', '') for name in db_files]
```

```python
for name, filename in zip(names, db_files):
    conn = db.OpsimDatabase(filename)
    
    bundleList = []
    sql = 'night > %f and night < %f and note not like "DD%%"' % ((365.25*3.5), (365.25*4.5) )
    metric = metrics.CountMetric('filter')
    slicer = slicers.HealpixSlicer()
    summaryStats = []
    plotDict = {'xlabel': 'Number of Observations', 'title':name, 'colorMin':0, 'colorMax': 200}
    bundleList.append(metricBundles.MetricBundle(metric,slicer,sql,
                                                 plotDict=plotDict,
                                                 summaryMetrics=summaryStats,
                                                 runName=name))
    bd = metricBundles.makeBundlesDictFromList(bundleList)
    bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outdir, resultsDb=resultsDb)
    bg.runAll()
    bg.plotAll(closefigs=False)
    conn.close()
```
